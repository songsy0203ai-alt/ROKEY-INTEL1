### ğŸ“‚ 2. ë°ì´í„°ì…‹ ê²½ë¡œ ì„¤ì • ë° YAML íŒŒì¼ ìƒì„±

```python
data_dir = '/home/rokey/ssy_ws/my_data/'
yaml_content = f"""
path: {data_dir}
train: train/images
val: valid/images
test: test/images

names:
  0: object  # ì‹¤ì œ í´ë˜ìŠ¤ëª…ìœ¼ë¡œ ìˆ˜ì •í•˜ì—¬ ì‚¬ìš©í•˜ì„¸ìš”.
"""

with open('local_data.yaml', 'w') as f:
    f.write(yaml_content)
print("ğŸ“ local_data.yaml ìƒì„± ì™„ë£Œ")
```

### ğŸ¤– 3. YOLOv8n ëª¨ë¸ ë¡œë“œ

```python
# ì •ìƒ ê²°ê³¼: 'yolov8n.pt' ê°€ì¤‘ì¹˜ íŒŒì¼ ë¡œë“œ ë©”ì‹œì§€ê°€ ì¶œë ¥ë©ë‹ˆë‹¤.
model = YOLO('yolov8n.pt')
```
